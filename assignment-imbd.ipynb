* considerare la lista di episodi di **Game of Thrones**, che possono essere reperiti in https://www.imdb.com/title/tt0944947/episodes/
* per ogni episodio viene riportata una descrizione, la valutazione media
* gli episodi sono distribuiti su più pagine, per cui serve un processo di **crawling**
* produrre dei grafici che riportino statistiche legate agli episodi, as esempio un bar plor che riporti il rating medio di ogni episodio

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
base_url = "https://www.imdb.com"
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 ' + ' (KHTML, like Gecko) Chrome/61.0.3163.100Safari/537.36'}

#dati interessanti
episodes_data = []

base_request = requests.get(base_url+"/title/tt0944947/episodes", headers=header)
soup = bs(base_request.text , "html.parser")

# prendo il primo episodio
ep = soup.find("a", attrs={'class':'ipc-title-link-wrapper'})
url = base_url+ep['href']

while url:
  print(f'Scraping url ({url})')
  #get data and save it in episodes_data
  r = requests.get(url , headers=header)
  p = bs(r.text , "html.parser")

  ep_szn = p.select(".sc-11bf8697-0")[0].text
  season = ep_szn.split('.')[0][1::]
  episode = ep_szn.split('.')[1][1::]

  rating = p.select("div.sc-3a4309f8-0:nth-child(1) > div:nth-child(1) > div:nth-child(1) > a:nth-child(2) > span:nth-child(1) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > span:nth-child(1)")[0].text

  n_reviews = p.select("div.sc-3a4309f8-0:nth-child(1) > div:nth-child(1) > div:nth-child(1) > a:nth-child(2) > span:nth-child(1) > div:nth-child(1) > div:nth-child(2) > div:nth-child(3)")[0].text

  director = p.select("ul.ipc-metadata-list:nth-child(3) > li:nth-child(1) > div:nth-child(2) > ul:nth-child(1) > li:nth-child(1) > a:nth-child(1)")[0].text


  runtime = p.select(".ipc-metadata-list--compact > li:nth-child(1) > div:nth-child(2)")[0].text

  episodes_data.append({
    'season': season ,
    'episode': episode ,
    'rating':rating ,
    'n_reviews': n_reviews ,
    'director':director ,
    'runtime' : runtime
  })
  print(episodes_data[len(episodes_data)-1])
  #get next episode link if it exists
  #return url=None if it doesnt
  try :
    anchor = p.select("a.ipc-icon-button:nth-child(5)")[0]['href']
    url = base_url+anchor
  except:
    url = None
```

### Convert output into json file per salvarlo

```python
import json
with open('got_episodes_data.json', 'w') as file:
  json.dump(episodes_data, file, indent=4)
```

#### Durata degli episodi

```python
def time_to_minutes(time_str):
    # Split the string by spaces
    parts = time_str.split()

    # Initialize variables to hold hours and minutes
    hours = 0
    minutes = 0

    # Iterate through the parts and extract hours and minutes
    for i in range(0, len(parts), 2):
        if parts[i + 1] == 'hour' or parts[i + 1] == 'hours':
            hours += int(parts[i])
        elif parts[i + 1] == 'minute' or parts[i + 1] == 'minutes':
            minutes += int(parts[i])

    # Convert hours to minutes and add to the total
    total_minutes = hours * 60 + minutes

    return total_minutes
```

```python
rdf = df['runtime'].apply(time_to_minutes)
rdf.plot(kind='hist',y='runtime', bins=35 ,title="Durata degli episodi in minuti", xlabel="Durata episodio", ylabel="Numero di episodi")
```
