```python
import os
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("../datasets/DataSets/netflix.csv", parse_dates=["date_added"])
df
```

* Trovare tutte le righe con diretctor "Robert Rodriguez" e tipo "Movie".

```python
f1 = df[(df["director"] == "Robert Rodriguez") & (df['type'] == "Movie")]
f1
```

* Trovare tutte le righe com data "2019-07-31" o direttore "Robert Altman".

```python
f2 = df[(df['date_added'] == "2019-07-31") | (df["director"] == "Robert Altman")]
f2
```

* Trovare tutte le righe con direttore "Orson Welles", "Aditya Kripalani", o "Sam Raimi".

```python
f3 = df[ df['director'].isin(["Orson Welles", "Aditya Kripalani","Sam Raimi"])]
f3
```

* Trovare tutte le righe in cui la data è tra May 1, 2019 e June 1, 2019.

```python
f4 = df[ (df["date_added"] >= "2019-05-01") & (df["date_added"] < "2019-06-01") ]
f4
```

```python
df_no_null = df[ df['director'].notnull()]
df_no_null
```
